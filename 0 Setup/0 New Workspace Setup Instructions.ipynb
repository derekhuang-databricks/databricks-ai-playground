# New workspace instructions

## Create a catalog

- Create a new catalog that is binded to ADLS Gen 2 cloud storage that everyone can work off

- Change the default catalog in the workspace to this newly created catalog under: `Settings` > `Advanced` > `Default catalog for the workspace` option

## Enable previews that support Agent Bricks

- Enable the following previews in the workspace under `Previews` for:

    - AI ParseDocument

    - External Tool Calling for Agents

    - Synthetic Agent Evaluation

    - Genie Research Agent

    - Managed MCP Servers

    - Managed MLflow Prompt Registry

    - Mosaic AI Agent Bricks Preview

    - Production Monitoring for MLflow

    - Agent Framework: On-Behalf-Of-User Authorization

    - Databricks Assistant Agent Mode
