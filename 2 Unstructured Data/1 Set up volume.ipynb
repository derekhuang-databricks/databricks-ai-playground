# Set up agent for querying unstructured data

Find a bunch of related PDFs online that may be relevant to a specific use case you're currently driving or looking to consider in future. For example; financial statements, public reports.

We will now use [Agent Bricks Knowledge Assistant](https://docs.databricks.com/aws/en/generative-ai/agent-bricks/knowledge-assistant) to create a RAG application without knowing all the technical knobs and whistles. Let's see how we can do that by querying unstructured data on the Databricks platform.

## Exercise I: Source your PDF data

Here are some examples of documents you can source into Databricks via a Volume:

- **Product**: technical specifications, product reviews

- **Marketing**: Marketing Intelligence reports (internal or exterior competitor analysis) conducted by research agencies

- **Manufacturing**: Summary of maintenance reports conducted by external parties (e.g. regulators)


## Exercise II: Upload Assets to Volume

1. Within the schema, create a "Volume" asset.

2. Within the Volume, optionally create a directory and upload your PDF documents. The [file size limit for Agent Bricks Knowledge Assistant](https://docs.databricks.com/aws/en/generative-ai/agent-bricks/knowledge-assistant#limitations) PDF documents is currently 50MB.

## Exercise III: Create Agent Bricks Knowledge Assistant

- Select the Knowledge Assistant brick.

- Use natural language to specify what you want the agent to do.

- Once created, the Knowledge Assistant will be available as a Model Serving endpoint. Note that in the first run, this could take some time while it creates the underlying RAG architecture.
