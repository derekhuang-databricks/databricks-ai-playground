# Set up Multi Agent architecture

You can use Agent Bricks Multi Agent Supervisor to bring your Genie spaces and Knowledge Assistant agents together into a seamless experience.

# Final Remarks

The Agent Bricks Knowledge Assistant agent allows you to build at speed, but doesn't cater for specific customisation at this stage. If you wanted to customise certain aspects of the Agent (e.g. chunking algorithm, selecting a specific embedding and retrieval model other than the specific `databricks-gte-large-en` embedding model), we would recommend use of the Agent Framework.
